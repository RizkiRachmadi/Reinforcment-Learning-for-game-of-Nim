```
#Making hierarchy using anytree
a = Node(0.5,foo='1')
b = Node(0.5, parent=a,foo='2')
c = Node(0.5, parent=a,foo='3')
d = Node(0.5, parent=a,foo='4')
e = Node(0.5, parent=b,foo='5')
f = Node(0.5, parent=b,foo='6')
g = Node(0.5, parent=c,foo='7')
h = Node(0.5, parent=c,foo='8')
i = Node(0.5, parent=d,foo='9')
j = Node(0.5, parent=d,foo='10')

k = Node(0.5, parent=e,foo='11')
l = Node(0.5, parent=e,foo='12')
m = Node(0.5, parent=f,foo='13')
n = Node(0.5, parent=g,foo='14')
o = Node(1, parent=h,foo='15')
p = Node(0.5, parent=i,foo='16')
q = Node(0.5, parent=j,foo='17')

r = Node(0.5, parent=k,foo='18')
s = Node(0, parent=l,foo='19')
t = Node(0, parent=m,foo='20')
u = Node(0, parent=n,foo='21')
v = Node(0, parent=p,foo='22')
w = Node(0, parent=q,foo='24')

x= Node(1, parent=r,foo='23')
```

```
#Checking Hierarchy
for pre, fill, node in RenderTree(a):
  print("%s%s" % (pre, node.name))
```

```
#Making self playing agent (1st move)
#Setting Starting point and parameter agent (exploration type)
path = ['1']
max = 0
list_node=[]
win = 0
lose = 0
startpoint = a
alpha = 0.1
explore_rate = 0.8
for x in range(0,10000):
  print("Iteration number",x)
  for y in range(1,(a.height+1)):
    #Rhe agent start first
    if (y%2) != 0:
      p = random.random()
      #Exploitation rate
      if (p <= explore_rate):
        indexx = random.randint(0,(len(startpoint.children)-1))
        path.append(startpoint.children[indexx].foo)
        nextpoint = find_by_attr(startpoint, name="foo", value=startpoint.children[indexx].foo)
        startpoint.name = startpoint.name + (alpha*(nextpoint.name - startpoint.name))
        startpoint = nextpoint
        if len(startpoint.children) == 0:
          startpoint = a
          print(path)
          path.clear()
          path = ['1']
          break
        else:
          pass
      #Exploration rate    
      elif (p>explore_rate):
        indexx = 0 
        for node in (startpoint.children):
          list_node.append(node.name)
        for aa in range(len(list_node)):
          if list_node[aa] > max:
            max = list_node[aa]
            indexx == aa
          else:
            pass
        list_node.clear()
        path.append(startpoint.children[indexx].foo)
        nextpoint = find_by_attr(startpoint, name="foo", value=startpoint.children[indexx].foo)
        startpoint.name = startpoint.name + (alpha*(nextpoint.name - startpoint.name))
        startpoint = nextpoint
        #checking if there is no longer children node from current node
        if len(startpoint.children) == 0:
          startpoint = a
          print(path)
          path.clear()
          path = ['1']
          break
        else:
          pass
      else:
        print("theres error in p")
    
    #random choice for opponent turn
    else:
      indexx = random.randint(0,(len(startpoint.children)-1))
      path.append(startpoint.children[indexx].foo)
      nextpoint = find_by_attr(startpoint, name="foo", value=startpoint.children[indexx].foo)
      startpoint.name = startpoint.name + (alpha*(nextpoint.name - startpoint.name))
      startpoint = nextpoint
      if len(startpoint.children) == 0:
        startpoint = a
        print(path)
        path = ['1']
        break
      else:
        pass
```
